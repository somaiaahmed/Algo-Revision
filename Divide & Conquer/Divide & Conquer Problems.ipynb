```python
def binary_search(K, q):
    left, right = 0, len(K) - 1
    while left <= right:
        middle = (left + right) // 2
        if K[middle] == q:
            return middle
        elif K[middle] < q:
            left = middle + 1
        else:
            right = middle - 1
    return -1

def multiple_search(K, Q):
    results = []
    for q in Q:
        results.append(binary_search(K, q))
    return results

# Example Input
n = 7
K = [1, 3, 7, 8, 9, 12, 15]
m = 4
Q = [9, 2, 15, 3]

# Example Output
print(multiple_search(K, Q))  # Output: [4, -1, 6, 1]

# Example Input
n = 5
K = [1, 5, 8, 12, 13]
m = 5
Q = [8, 1, 23, 1, 11]

# Example Output
print(multiple_search(K, Q))  # Output: [2, 0, -1, 0, -1]
```

```python
def majority_element(nums):
    n = len(nums)
    
    # Phase 1: Boyer-Moore Voting Algorithm to find the candidate
    candidate = None
    count = 0
    
    for num in nums:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1
    
    # Phase 2: Validate the candidate
    count = 0
    for num in nums:
        if num == candidate:
            count += 1
    
    # Return 1 if the candidate appears more than n//2 times, otherwise 0
    if count > n // 2:
        return 1
    else:
        return 0

# Example usage:
# Sample 1
nums1 = [2, 3, 9, 2, 2]
print(majority_element(nums1))  # Output: 1

# Sample 2
nums2 = [1, 2, 3, 1]
print(majority_element(nums2))  # Output: 0
```

```python
import random

def randomized_quick_sort(arr):
    def partition(arr, low, high):
        pivot = arr[high]  # Choose the last element as the pivot
        lt = low  # pointer for elements less than pivot
        gt = high  # pointer for elements greater than pivot
        i = low  # pointer for traversing the array

        while i <= gt:
            if arr[i] < pivot:
                arr[lt], arr[i] = arr[i], arr[lt]
                lt += 1
                i += 1
            elif arr[i] > pivot:
                arr[gt], arr[i] = arr[i], arr[gt]
                gt -= 1
            else:  # arr[i] == pivot
                i += 1

        return lt, gt  # Return the indices that separate the low, equal, and high sections

    def quick_sort(arr, low, high):
        if low < high:
            # Randomly select a pivot and move it to the end for partitioning
            pivot_index = random.randint(low, high)
            arr[pivot_index], arr[high] = arr[high], arr[pivot_index]
            
            # Partition the array around the pivot
            lt, gt = partition(arr, low, high)
            
            # Recursively sort the subarrays
            quick_sort(arr, low, lt - 1)  # Sort the low part
            quick_sort(arr, gt + 1, high)  # Sort the high part

    quick_sort(arr, 0, len(arr) - 1)
    return arr

# Example usage:
arr = [2, 3, 9, 2, 2]
print("Sorted array:", randomized_quick_sort(arr))
```

```python
def merge_and_count(arr, temp_arr, left, right):
    if left == right:
        return 0  # Base case: no inversions in a single element

    mid = (left + right) // 2
    inv_count = 0

    # Count inversions in the left and right halves
    inv_count += merge_and_count(arr, temp_arr, left, mid)
    inv_count += merge_and_count(arr, temp_arr, mid + 1, right)

    # Count inversions while merging the two halves
    inv_count += merge(arr, temp_arr, left, mid, right)

    return inv_count

def merge(arr, temp_arr, left, mid, right):
    i = left  # Starting index for left subarray
    j = mid + 1  # Starting index for right subarray
    k = left  # Starting index to be sorted
    inv_count = 0

    # Merge the two subarrays while counting inversions
    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp_arr[k] = arr[i]
            i += 1
        else:
            temp_arr[k] = arr[j]
            inv_count += (mid - i + 1)  # All remaining elements in the left subarray are inversions
            j += 1
        k += 1

    # Copy the remaining elements of left subarray, if any
    while i <= mid:
        temp_arr[k] = arr[i]
        i += 1
        k += 1

    # Copy the remaining elements of right subarray, if any
    while j <= right:
        temp_arr[k] = arr[j]
        j += 1
        k += 1

    # Copy the sorted subarray into the original array
    for i in range(left, right + 1):
        arr[i] = temp_arr[i]

    return inv_count

def count_inversions(arr):
    n = len(arr)
    temp_arr = [0] * n
    return merge_and_count(arr, temp_arr, 0, n - 1)

# Example usage:
arr = [2, 3, 9, 2, 9]
print("Number of inversions:", count_inversions(arr))
```

```python
import bisect

def count_segments_containing_points(segments, points):
    # Step 1: Sort segments by start (li) and end (ri)
    segment_starts = sorted([l for l, r in segments])
    segment_ends = sorted([r for l, r in segments])
    
    # Step 2: Sort points along with their original indices to return the results in the input order
    indexed_points = sorted([(p, i) for i, p in enumerate(points)])
    
    print(f"indexed_points = {indexed_points}")
    
    # Step 3: Array to store the number of segments containing each point
    result = [0] * len(points)
    
    active_segments = 0
    segment_idx = 0
    
    # Step 4: Process each point
    for point, original_idx in indexed_points:
        # Add segments that start before or at this point
        while segment_idx < len(segment_starts) and segment_starts[segment_idx] <= point:
            active_segments += 1
            segment_idx += 1
        
        # Remove segments that end before this point
        while len(segment_ends) > 0 and segment_ends[0] < point:
            segment_ends.pop(0)
            active_segments -= 1
        
        # Store the number of active segments containing this point
        result[original_idx] = active_segments

    return result

# Example
n, m = 2,3
segments = [(0, 5), (7, 10)]
points = [1, 6, 11]
# Compute the result
result = count_segments_containing_points(segments, points)

# Output the result
print(" ".join(map(str, result)))

# Example
n, m = 1,3
segments = [(-10, 10)]
points = [-100, 100, 0]
# Compute the result
result = count_segments_containing_points(segments, points)

# Output the result
print(" ".join(map(str, result)))

# Example
n, m = 3, 2
segments = [(0, 5), (-3, 2), (7, 10)]
points = [1, 6]
# Compute the result
result = count_segments_containing_points(segments, points)

# Output the result
print(" ".join(map(str, result)))
```

```python
import math

def closest_pair(points):
    def dist(p1, p2):
        return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)
    
    def closest_pair_rec(points_x, points_y):
        if len(points_x) <= 3:
            # Brute force for small number of points
            min_dist = float('inf')
            for i in range(len(points_x)):
                for j in range(i + 1, len(points_x)):
                    min_dist = min(min_dist, dist(points_x[i], points_x[j]))
            return min_dist
        
        mid = len(points_x) // 2
        mid_x = points_x[mid][0]
        
        # Split into two halves
        left_x = points_x[:mid]
        right_x = points_x[mid:]
        
        # Sort points by y-coordinate in both halves
        left_y = [p for p in points_y if p[0] <= mid_x]
        right_y = [p for p in points_y if p[0] > mid_x]
        
        # Recursively find the closest pair in both halves
        d1 = closest_pair_rec(left_x, left_y)
        d2 = closest_pair_rec(right_x, right_y)
        
        # Get the minimum distance from both sides
        d = min(d1, d2)
        
        # Check the points in the strip
        strip = [p for p in points_y if abs(p[0] - mid_x) < d]
        
        # Compare points in the strip
        min_d_strip = d
        for i in range(len(strip)):
            for j in range(i + 1, len(strip)):
                if (strip[j][1] - strip[i][1]) >= min_d_strip:
                    break
                min_d_strip = min(min_d_strip, dist(strip[i], strip[j]))
        
        return min(min_d_strip, d)
    
    # Sort points by x and y
    points_x = sorted(points, key=lambda p: p[0])
    points_y = sorted(points, key=lambda p: p[1])
    
    # Call the recursive function
    return closest_pair_rec(points_x, points_y)

# Examole
n = 2
points = [(0, 0), (3, 4)]

# Call the function and output the result
result = closest_pair(points)
print(f"{result:.6f}")

# Examole
n = 11
points = [(4, 4), (-2, -2), (-3, -4), 
          (-1, 3), (2, 3), (-4, 0), (1, 1),
          (-1, -1), (3, -1), (-4, 2), (-2, 4)]

# Call the function and output the result
result = closest_pair(points)
print(f"{result:.6f}")
```
